    1. Trangular Number Sequence-1

This code generates a triangular number pattern where each row starts from 1 and ends at the current row number. The outer loop controls the rows, and the inner loop prints numbers in increasing order.




```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j,end=" ")
    
    print()
```

     2. Trangular Number Pattern-2

This code creates a pattern where each row starts from the current row number and ends at 5. The outer loop iterates in reverse, and the inner loop prints numbers in increasing order.



```python
for i in range(5,0,-1):
    for j in range(i,6):
        print(j,end=" ")
    print()
```

3. Trangular Number Pattern-3

This code generates a pattern where each row contains numbers in decreasing order, starting from the current row number down to 1. The outer loop controls the rows, and the inner loop prints numbers in reverse.




```python
for i in range(1,6):
    for j in range(i,0,-1):
        print(j,end=" ")
    print()     
```

4.     Trangle Number Pattern - 4

This code creates a pattern where each row contains numbers in decreasing order, starting from 5 and reducing by one with each column. The outer loop iterates in reverse, and the inner loop prints numbers in reverse order.





```python
for i in range(5,0,-1):
    for j in range(5,5-i,-1):
        print(j,end=" ")
    print()
```

7.     Triangle Star Pattern

This code generates a star pattern where each row contains a decreasing number of stars. The outer loop iterates in reverse, and the number of stars printed decreases with each iteration.




```python
for i in range(5,0,-1):
    print("*"*i)
```

8.     Triangle Number Pattern 6

This code generates a diamond-like pattern. The first half creates an increasing triangular pattern, and the second half creates a decreasing triangular pattern. Two loops are used for each half.



```python
# Half of the Triangle
for i in range(1,6):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
# other half of the Triangle
for i in range(4,0,-1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```
