    14. Triangular Number Pattern




```python
x = 5
for i in range(1,6):
    for j in range(i):
        print(x , end=" ")
    x -= 1
    print()
```

    15.Triangular number pattern 

 

```python
for i in range(1,6):
    for j in range(6-i):
        print(i,end=" ")
    print()
```

    16. Tringular Number pattern



```python
# Half Pattern
for i in range(5,0,-1):
    for j in range(6-i):
        print(i,end=" ")
    print()
# Remaining Pattern
for i in range(2,6):
    for j in range(6-i):
        print(i,end=" ")
    print()
```
